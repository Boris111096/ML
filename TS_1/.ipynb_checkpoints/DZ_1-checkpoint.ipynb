```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import statsmodels.tsa.api as smt
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from scipy.stats import boxcox
%matplotlib inline
```

```python
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
def tsplot(y, lags=None, figsize=(14, 8), style='bmh', series_name = ''):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title(series_name + ' — Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)

        plt.tight_layout()
    return
```

```python
sales_of_company_x = pd.read_csv("Series/monthly-sales-of-company-x-jan-6.csv")
robberies_in_boston = pd.read_csv("Series/monthly-boston-armed-robberies-j.csv")
airlines_passengers = pd.read_csv("Series/international-airline-passengers.csv")
mean_monthly_temp = pd.read_csv("Series/mean-monthly-air-temperature-deg.csv")
dowjones_closing = pd.read_csv("Series/weekly-closings-of-the-dowjones-.csv")
female_births = pd.read_csv("Series/daily-total-female-births-in-cal.csv")
```

```python
all_series = {
    "Monthly sales of company X": sales_of_company_x["Count"],
    "Monthly Boston armed robberies": robberies_in_boston["Count"],
    "International airline passengers: monthly totals in thousands": airlines_passengers["Count"],
    "Mean monthly air temperature (Deg. F) Nottingham Castle": mean_monthly_temp["Deg"],
    "Weekly closings of the Dow-Jones industrial average": dowjones_closing["Close"],
    "Daily total female births in California": female_births["Count"]
}
```

## 1 шаг

```python
for i, key in enumerate(all_series.keys()):
    print(f'{i}) {key}')
    tsplot(all_series[key], series_name= f'{i}) {key}')
```

## 2 шаг

```python
for i, key in enumerate(all_series.keys()):
    print(f'{i}) {key}')
    all_series[key] = boxcox(all_series[key], 0)
    tsplot(all_series[key], series_name=f'{i}) {key}')
```

## 3 шаг

```python
for i, key in enumerate(all_series.keys()):
    print(f'{i}) {key}')
    all_series[key] = all_series[key][1:] - all_series[key][:-1]
    tsplot(all_series[key], series_name=f'{i}) {key}')
```

## 4 шаг

```python
for i, key in enumerate(all_series.keys()):
    print(f'{i}) {key}')
    all_series[key] = all_series[key][12:] - all_series[key][:-12]
    tsplot(all_series[key], series_name=f'{i}) {key}')
```

Итак, я прогнал все таймсерии по одному шаблону, и теперь можно сделать следующие выводы:

# Итог

1. "Mean monthly air temperature (Deg. F) Nottingham Castle" (№ 3) и "Daily total female births in California" (№ 5) - изначально были стационарные (1 шаг). Их можно не обрабатывать.
2. "Weekly closings of the Dow-Jones industrial average" (№ 4) прошел тест и выдал нормальный график на 3 шаге.
3. "Monthly Boston armed robberies" (№ 1) и "International airline passengers: monthly totals in thousands" (№ 2) - на 4 шаге
4. "Monthly sales of company X" (№ 0) - самый проблемный получился. На 3 шаге он прошел тест, но график получился неудовлетворительным. Там надо просто по-другому обработать сезонность на 4 шаге:

```python
bad_series = sales_of_company_x["Count"].copy()
bad_series = boxcox(bad_series, 0)
bad_series = bad_series[1:] - bad_series[:-1]
bad_series = bad_series[12:] - bad_series[:-12]

tsplot(bad_series, series_name=f'0) Monthly sales of company X')
```

```python
bad_series = sales_of_company_x["Count"].copy()
bad_series = boxcox(bad_series, 0)
bad_series = bad_series[1:] - bad_series[:-1]
bad_series = bad_series[24:] - bad_series[:-24]

tsplot(bad_series, series_name=f'0) Monthly sales of company X')
```
