```python
# уберем наны в таргете, которые нам  предлагают предсказать

df = data[~data.shot_made_flag.isna()].copy()
test_df = data[data.shot_made_flag.isna()].copy()
df
```

```python
# проверим, что все значения не пустые

df.info()
```

```python
# loc_x, loc_y, lat, lon - это location_x, location_y, latitude and longitude

alpha = 0.02
plt.figure(figsize=(10,10))


plt.subplot(121)
plt.scatter(df.loc_x, df.loc_y, color='blue', alpha=alpha)
plt.title('loc_x and loc_y')


plt.subplot(122)
plt.scatter(df.lon, df.lat, color='green', alpha=alpha)
plt.title('lat and lon')
```

```python
# переведём эти признаки в полярные координаты

df.loc[:,'dist'] = np.sqrt(df.loc[:,'loc_x']**2 + df.loc[:,'loc_y']**2)

loc_x_zero = df['loc_x'] == 0

df.loc[:,'angle'] = np.array([0]*len(df))
df.loc[~loc_x_zero,'angle'] = np.arctan(df.loc[~loc_x_zero,'loc_y'] / df.loc[~loc_x_zero,'loc_x'])
df.loc[loc_x_zero,'angle'] = np.pi / 2 # обрабатываем ошибки деления на 0
```

```python
df[['minutes_remaining','seconds_remaining']]
```

```python
# объединяем minutes_remaining и seconds_remaining

df['remaining_time'] = df['minutes_remaining'] * 60 + df['seconds_remaining']
```

```python
df.season
```

```python
# Season объединяет в себе год и номер сезона. Оставим только номер сезона.

df['season'] = df['season'].apply(lambda x: int(x.split('-')[1]) )
df['season'].unique()
```

```python
# Бесполезные столбцы

print(df['team_id'].unique())
print(df['team_name'].unique())
```

```python
# одна и та же информация. можно один столбец убрать

pd.DataFrame({'matchup':df.matchup, 'opponent':df.opponent})
```

```python
# мы вывели уже такой же столбец - dist. следовательно от shot_distance можно избавиться

plt.figure(figsize=(5,5))

plt.scatter(df.dist, df.shot_distance, color='blue')
plt.title('dist and shot_distance')
```

```python
# shot_zone_area, shot_zone_basic, shot_zone_range - просто зоны бросков. это уже опять же учитано в полярных координатах,
# так что еще раз они нам не нужны 

import matplotlib.cm as cm
plt.figure(figsize=(20,10))

def scatter_plot_by_category(feat):
    alpha = 0.1
    gs = df.groupby(feat)
    cs = cm.rainbow(np.linspace(0, 1, len(gs)))
    for g, c in zip(gs, cs):
        plt.scatter(g[1].loc_x, g[1].loc_y, color=c, alpha=alpha)

# shot_zone_area
plt.subplot(131)
scatter_plot_by_category('shot_zone_area')
plt.title('shot_zone_area')

# shot_zone_basic
plt.subplot(132)
scatter_plot_by_category('shot_zone_basic')
plt.title('shot_zone_basic')

# shot_zone_range
plt.subplot(133)
scatter_plot_by_category('shot_zone_range')
plt.title('shot_zone_range')
```

```python
# дропаем ненужные столбцы

drops = ['shot_id', 'team_id', 'team_name', 'shot_zone_area', 'shot_zone_range', 'shot_zone_basic',
         'matchup', 'lon', 'lat', 'seconds_remaining', 'minutes_remaining',
         'shot_distance', 'loc_x', 'loc_y', 'game_event_id', 'game_id', 'game_date']
df = df.drop(columns=drops)
df
```

```python
# обрабатываем категориальные переменные

categorical_vars = ['action_type', 'combined_shot_type', 'shot_type', 'opponent', 'period', 'season']

df = pd.get_dummies(df, columns = categorical_vars)
df
```

```python
rf = RandomForestRegressor(random_state=0)

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='shot_made_flag')
                                                    , df.shot_made_flag
                                                    , test_size=0.33
                                                    , random_state=42)

rf.fit(X_train, y_train)
print(f'train_score: {rf.score(X_train, y_train)}')
y_pred = rf.predict(X_test)
print(f'test_score: {r2_score(y_test, y_pred)}')
features = pd.DataFrame([df.drop(columns='shot_made_flag').columns, rf.feature_importances_]).T.sort_values(1, ascending = False)
features.columns = ['feature','score']
features = features.head(10).sort_values('score')
plt.barh(np.arange(len(features)), features.score)
plt.yticks(np.arange(len(features)),features.feature)
```

Явное переобучение)

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='shot_made_flag')
                                                    , df.shot_made_flag
                                                    , test_size=0.33
                                                    , random_state=42)
scaler = StandardScaler()

lr = LogisticRegression().fit(scaler.fit_transform(X_train), y_train)

y_pred = lr.predict(scaler.fit_transform(X_test))
print(f'train_score: {lr.score(scaler.fit_transform(X_train), y_train)}')
print(f'test_score: {roc_auc_score(y_test, y_pred)}')
```

```python
rr = RidgeClassifier().fit(scaler.fit_transform(X_train), y_train)

y_pred = rr.predict(scaler.fit_transform(X_test))
print(f'train_score: {rr.score(scaler.fit_transform(X_train), y_train)}')
print(f'test_score: {roc_auc_score(y_test, y_pred)}')
```

```python
la = LassoCV(cv=5, random_state=0).fit(scaler.fit_transform(X_train), y_train)

y_pred = la.predict(scaler.fit_transform(X_test))

print(f'train_score: {la.score(scaler.fit_transform(X_train), y_train)}')
print(f'test_score: {roc_auc_score(y_test, y_pred)}')
```

```python
features = pd.DataFrame(list(zip(X_test.columns, la.coef_)))#.sort_values(1, ascending = False)
features = pd.merge(features, pd.DataFrame(list(zip(X_test.columns, rr.coef_[0]))), how = 'inner', on = 0)
features = pd.merge(features, pd.DataFrame(list(zip(X_test.columns, lr.coef_[0]))), how = 'inner', on = 0)
features.columns = ['feature','la_coeff','rr_coeff','lr_coeff']
features = features.sort_values(['la_coeff','rr_coeff','lr_coeff'], ascending = False)
features
```

```python
plt.figure(figsize=(20,10))

len_ = np.arange(len((features.feature)))

plt.scatter(features.index, features.la_coeff)
plt.scatter(features.index, features.rr_coeff)
plt.scatter(features.index, features.lr_coeff)
plt.title('values of features coeff')
print(f'total features: {len(features)}')
plt.show()
```
