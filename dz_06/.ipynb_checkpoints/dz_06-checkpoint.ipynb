```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,5)
plt.rcParams["font.size"] = 15

import seaborn as sns
sns.set_theme(style="ticks")

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, normalize
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.datasets import load_boston
data = pd.read_csv(load_boston()['filename'], skiprows=1)
```

```python
data
```

```python
# попробуем построить модель

clf = DecisionTreeClassifier(max_depth=10)

clf.fit(data.drop(columns='RAD'), data.RAD)
y_pred = clf.predict(data.drop(columns='RAD'), data.RAD.to_list())
accuracy_score(data.RAD.to_list(), y_pred)
```

```python
# вышло неплохо, так что можно посмотреть топовые фичи

plt.barh(np.arange(len(clf.feature_importances_)), clf.feature_importances_)
plt.yticks(np.arange(len(data.drop(columns='RAD').columns)),data.drop(columns='RAD').columns)
```

```python
# возьмем фичи, к примеру, > 5%

features_df = pd.DataFrame([data.drop(columns='RAD').columns, clf.feature_importances_]).T.sort_values(1, ascending=False)
features_df.columns = ["feature","rate"]
features_df
```

```python
# дропнем слабые фичи и прогоним модель на разную глубину max_depth = n

best_score = 0
score_list = {'train_score':[],'test_score':[]}
n = range(3,30)

X_train, X_test, y_train, y_test = train_test_split(data[features_df[features_df.rate>0.05].feature.to_list()]
                                                    , data.RAD.to_list()
                                                    , test_size=0.25
                                                    , random_state=42)

for i in n:
    clf = DecisionTreeClassifier(max_depth=i)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    score_list['train_score'].append(clf.score(X_train, y_train))
    score_list['test_score'].append(accuracy_score(y_test,y_pred))
    if best_score < accuracy_score(y_test,y_pred):
        max_depth = i
        best_score = accuracy_score(y_test,y_pred)

print(f"best_score = {best_score}\nmax_depth = {max_depth}")

plt.plot(np.array(n), score_list['train_score'])
plt.plot(np.array(n), score_list['test_score'])
plt.xticks(np.array(n))
plt.show()
```

Как видно, рост качества модели на тестовых данных заканчивается на глубине 13. 

```python
from sklearn.tree import export_graphviz

def get_tree_dot_view(clf, feature_names=None, class_names=None):
    print(export_graphviz(clf, out_file=None, filled=True, feature_names=feature_names, class_names=class_names))
    
clf = DecisionTreeClassifier(max_depth=13)
clf.fit(X_train, y_train)
get_tree_dot_view(clf)
```
