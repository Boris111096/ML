```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,5)
plt.rcParams["font.size"] = 15

import seaborn as sns
sns.set_theme(style="ticks")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, normalize
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import load_boston
data = pd.read_csv(load_boston()['filename'], skiprows=1)
```

```python
data
```

```python
print(load_boston()['DESCR'])
```

```python
# Находим количество уникальных значений каждого столбцы. Это позволит понять, для каких категориальных переменных можно
# применить get_dummies 

for i in data.columns:
    print(f'{i}\t  {len(data[i].unique())}')
```

```python
data.ZN.sort_values().unique()
```

```python
# Разбиваем только 2 стобца 'CHAS' и 'RAD'. Остальные столбцы можно рассматривать как непрерывные или дискретные величины

df = data.copy()
df[['CHAS','RAD']].astype('str')
df = pd.get_dummies(df, columns=['CHAS','RAD'])
df
```

```python
data.ZN.sort_values().unique()
```

```python
def norm(data, n=2):
    """
    Функция возвращает нормализованные данные полинома n-ой степени.
    """
    if n==1:
        return normalize(data)
    else:
        return PolynomialFeatures(n).fit_transform(normalize(data))
```

```python
# Пообучаем модель и потестируем на разных выборках

best_score = 0
test_score = []
pred_score = []

for n in range(1,36):
    X_train, X_test, y_train, y_test = train_test_split(norm(df.drop(columns='MEDV'), n=1)
                                                        , df.MEDV
                                                        , test_size=n/100
                                                        , random_state=42)



    lr = LinearRegression().fit(norm(X_train), y_train)
    y_pred = lr.predict(norm(X_test))
    test_score.append(lr.score(norm(X_train), y_train))
    pred_score.append(r2_score(y_test,y_pred))
#     print(f"Выборка {n/100}:")
#     print(f"\tОценка трейновых данных: {lr.score(norm(X_train), y_train)}\n\tОценка тестовых данных: {r2_score(y_test,y_pred)}")
    if best_score < r2_score(y_test,y_pred):
        best_score = r2_score(y_test,y_pred)
        best_test_size = n/100
print(f"best_test_size = {best_test_size}\nbest_score = {best_score}")
plt.plot([n for n in range(1,36)], test_score)
plt.plot([n for n in range(1,36)], pred_score)
plt.show()
```

```python
# А теперь без нормализатора

best_score = 0
test_score = []
pred_score = []

for n in range(1,36):
    X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='MEDV')
                                                        , df.MEDV
                                                        , test_size=n/100
                                                        , random_state=42)



    lr = LinearRegression().fit(norm(X_train), y_train)
    y_pred = lr.predict(norm(X_test))
    test_score.append(lr.score(norm(X_train), y_train))
    pred_score.append(r2_score(y_test,y_pred))
#     print(f"Выборка {n/100}:")
#     print(f"\tОценка трейновых данных: {lr.score(norm(X_train), y_train)}\n\tОценка тестовых данных: {r2_score(y_test,y_pred)}")
    if best_score < r2_score(y_test,y_pred):
        best_score = r2_score(y_test,y_pred)
        best_test_size = n/100
print(f"best_test_size = {best_test_size}\nbest_score = {best_score}")
plt.plot([n for n in range(1,36)], test_score)
plt.plot([n for n in range(1,36)], pred_score)
plt.show()
```

С нормализатором немного получше)  
В целом нужно больше данных.

```python
# Попробуем дропать по очереди все столбцы и смотреть, что получится)

best_score = 0
test_score = []
pred_score = []
best_score_in_columns_dict = {}

for c in df.columns:
    droped_df = df.copy()
    droped_df = droped_df.drop(columns = [c, 'MEDV'])
    best_score_in_columns = 0
    for n in range(1,36):
        X_train, X_test, y_train, y_test = train_test_split(norm(droped_df, n=1)
                                                            , df.MEDV
                                                            , test_size=n/100
                                                            , random_state=42)



        lr = LinearRegression().fit(norm(X_train), y_train)
        y_pred = lr.predict(norm(X_test))
        test_score.append(lr.score(norm(X_train), y_train))
        pred_score.append(r2_score(y_test,y_pred))
        
        if best_score < r2_score(y_test,y_pred):
            droped_col = c
            best_score = r2_score(y_test,y_pred)
            best_test_size = n/100
        if best_score_in_columns < r2_score(y_test,y_pred):
            best_score_in_columns = r2_score(y_test,y_pred)
    best_score_in_columns_dict[c] = best_score_in_columns
    
print(f"best_test_size = {best_test_size}\nbest_score = {best_score}\ndroped_col = {droped_col}")
```

Если дропать колонки, можно добиться лучшего результата)

```python
scores_df = pd.DataFrame([best_score_in_columns_dict[i] for i in best_score_in_columns_dict]
                         , [i for i in best_score_in_columns_dict]).sort_values(0, ascending = False).reset_index()
scores_df.columns = ['droped_column','score']
scores_df
```

```python
# А если дропнуть сразу несколько?

best_score = 0

for i in range(10):
    list_ = list(scores_df.loc[:i,'droped_column'])
    list_.append('MEDV')
    droped_df = df.copy()
    droped_df = droped_df.drop(columns = list_)
    X_train, X_test, y_train, y_test = train_test_split(norm(droped_df, n=1)
                                                        , df.MEDV
                                                        , test_size=0.01
                                                        , random_state=42)



    lr = LinearRegression().fit(norm(X_train), y_train)
    y_pred = lr.predict(norm(X_test))
    if best_score < r2_score(y_test,y_pred):
        droped_col = list(scores_df.loc[:i,'droped_column'])
        best_score = r2_score(y_test,y_pred)

print(f"best_score = {best_score}\ndroped_col = {droped_col}")
```
