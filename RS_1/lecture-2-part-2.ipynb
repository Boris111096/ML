```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

from tqdm.notebook import tqdm

from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.neighbors import NearestNeighbors

%matplotlib inline
```

```python
# links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
def change_string(s):
    return ' '.join(s.replace(' ', '').replace('-', '').split('|'))
```

```python
# tfidf по жанрам

temp_df = pd.DataFrame(TfidfTransformer().fit_transform(CountVectorizer().fit_transform(movies.genres.to_list())).toarray())

movies.loc[:, 'genres'] = movies.loc[:, 'genres'].apply(lambda x: change_string(x))
movies.loc[:, temp_df.columns.tolist()] = temp_df.loc[:,:]
movies
```

```python
# считаем метрики для рейтинга, проставленного пользователем

users_ratings = pd.DataFrame([ratings.groupby('userId').rating.mean()
                              , ratings.groupby('userId').rating.median()
                              , ratings.groupby('userId').rating.std()
                              , ratings.groupby('userId').rating.var()
                             ]
                            ).T.reset_index()
users_ratings.columns = ['userId', 'u_mean', 'u_median', 'u_std', 'u_var']
users_ratings
```

```python
# считаем метрики для рейтинга, проставленный фильмам

movies_ratings = pd.DataFrame([ratings.groupby('movieId').rating.mean()
                              , ratings.groupby('movieId').rating.median()
                              , ratings.groupby('movieId').rating.std()
                              , ratings.groupby('movieId').rating.var()
                             ]
                            ).T.reset_index()
movies_ratings.columns = ['movieId', 'm_mean', 'm_median', 'm_std', 'm_var']
movies_ratings
```

```python
# tfidf по тегам (уменьшаем размерность с 1400+ до 20 фичей)

from sklearn.decomposition import PCA

movies_tags = tags.groupby('movieId').tag.apply(lambda x: ' '.join([change_string(str(i)) for i in x.to_list()])).reset_index()

pca = PCA(n_components=20)
pca.fit(TfidfTransformer().fit_transform(CountVectorizer().fit_transform(movies_tags.tag.to_list())).toarray().T)

temp_df = pd.DataFrame(pca.components_.T)

movies_tags.loc[:, temp_df.columns.tolist()] = temp_df.loc[:,:]
movies_tags
```

```python
# соединяем датафреймы

resultdf = pd.merge(ratings
                    , movies
                    , on = 'movieId'
                   )

resultdf = pd.merge(resultdf
                    , users_ratings
                    , on = 'userId'
                   )

resultdf = pd.merge(resultdf
                    , movies_ratings
                    , on = 'movieId'
                   )

resultdf = pd.merge(resultdf
                    , movies_tags
                    , on = 'movieId'
                   )

resultdf
```

```python
# нанов мало, поэтому просто удаляем строки + дропаем столбцы
data = resultdf.iloc[:, 2:].copy()
data = data[data.notna().all(axis=1)][data.dtypes[(data.dtypes == float) | (data.dtypes == int)].index.tolist()]
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, 1:]
                                                    , data.rating
                                                    , test_size=0.2
                                                    , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = lr.predict(X_test)

mean_squared_error(y_test, y_pred)
```
